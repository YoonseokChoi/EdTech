```python
import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings("ignore")
```

```python
df = sns.load_dataset("penguins")

print(df.shape)
df.head()
```

```python
_= sns.histplot(
    df,
    x = 'flipper_length_mm',
#     bin = 10
)
```

```python
_= sns.histplot(
    df,
    x = 'flipper_length_mm',
    hue = 'species',
    multiple = 'stack',
    bins = 30
)
```

```python
_= sns.histplot(
    df,
    x = 'flipper_length_mm',
    hue = 'species',
    multiple = 'fill',
#     bins = 30
)
```

```python
_=sns.kdeplot(
    df,
    x = 'body_mass_g',
    hue = 'species',
#     fill = True
)
```

```python
_=sns.kdeplot(
    df,
    x = 'flipper_length_mm',
    hue = 'species',
    bw_adjust=0.8,
    fill = True
)
```

```python
_=sns.rugplot(
    data = df,
    x = 'flipper_length_mm',
    hue = 'species'
)
```

```python
_= sns.rugplot(
    data=df,
    x="flipper_length_mm",
    y="body_mass_g",
    hue="species",
)

plt.show()
```

```python
_=sns.displot(
    data=df,
    x="flipper_length_mm",
    y="body_mass_g",
    hue="species",
    kind="kde",
    rug=True,
)
```

```python
sns.displot(
    data=df,
    x="flipper_length_mm",
    y="body_mass_g",
    hue="species",
    kind="hist",
    rug=True,
)

plt.show()
```
