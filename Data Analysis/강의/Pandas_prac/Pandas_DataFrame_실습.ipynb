## API 활용

### 예제 1

```python
import requests

api_token = ""
movie_id = 912649

# URL
base_url = "https://api.themoviedb.org/3"
endpoint = f"/movie/{movie_id}"
url = base_url + endpoint

# Parameter
params = {'language': 'ko-KR'}
# params = {'query': 'Toy Story'}

# Header
headers = {
    "Authorization": f"Bearer {api_token}",
    "accept": "application/json"
}

# Request
response = requests.get(
    url, 
    params = params, 
    headers = headers
)

print(response)
```

```python
from pprint import pprint

res_data = response.json()
pprint(res_data)
```

```python
data = {
    'ID': [res_data['id']],
    '제목': [res_data['title']],
    '개요': [res_data['overview']],
    '개봉일자': [res_data['release_date']]
}

df = pd.DataFrame(data)
df
```

### 예제 2

```python
import requests

api_token = ""

# URL
base_url = "https://api.themoviedb.org/3"
endpoint = f"/movie/now_playing"
url = base_url + endpoint

# Parameter
params = {
    'language': 'ko-KR', 
    'page':'1'
}

# Header
headers = {
    "Authorization": f"Bearer {api_token}",
    "accept": "application/json"
}

# Request
response = requests.get(
    url, 
    params = params, 
    headers = headers
)

print(response)
```

```python
df = pd.DataFrame(response.json()['results'])
print(df.shape)
df.head()
```
