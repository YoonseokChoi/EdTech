# import

```python
import requests
from pprint import pprint
import pandas as pd
```

# Request

```python
airports = [
    "RKJJ",  # 광주공항
    "RKJB",  # 무안공항
    "RKSI",  # 인천공항
    "RKNY",  # 양양공항
    "RKPC",  # 제주공항
    "RKPK",  # 김해공항
    "RKTU",  # 청주공항
    "RKPU",  # 울산공항
    "RKJY",  # 여수공항
    "RKSS",  # 김포공항
    "RKTH",  # 포항공항
    "RKTN",  # 대구공항
    "RKPS"   # 사천공항
]

data = []
for airport in airports:
    url = 'http://apis.data.go.kr/1360000/AirInfoService/getAirInfo'
    params ={
        'serviceKey' : '',
        'numOfRows' : '100000', 
        'pageNo' : '1', 
        'dataType' : 'json', 
        'fctm' : '202412181400', 
        'icaoCode' : airport 
    }

    response = requests.get(url, params=params)
    print(response)
    data.extend(response.json()['response']['body']['items']['item'])
```

# DataFrame 생성

```python
pprint(data)
```

```python
import pandas as pd

df = pd.DataFrame(data)
df
```
