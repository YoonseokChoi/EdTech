# filter() 연습

```python
df['부서'].value_counts()
```

```python
# def greater_count(group):
#     # 직원 수가 2명 보다 많은 부서는 True 반환
#     return len(group) > 2

# df.groupby("부서").filter(greater_count)

df.groupby("부서").filter(lambda x : len(x)>2)
```

```python
df.groupby('부서')['연봉'].mean()
```

```python
df.groupby('부서').filter(lambda x : x['연봉'].mean()>=6000)
```

```python
df.groupby('부서')['나이'].mean()
```

```python
df.groupby('부서').filter(lambda x : x['나이'].mean() < 35)
```

# agg() 연습

```python
df['연봉'].agg(['mean', 'max', 'min'])
```

```python
df[['나이']].agg(['mean', 'max', 'min'])
```

```python
df[['연봉','나이']].agg(['mean', 'max', 'min']).T
```

```python
df.agg({
    "이름": "count",
    "나이": ["mean","min","max"],
    "연봉": ["median", "mean"]
})
```

```python
df3 = df.groupby(['지역']).agg({
    "이름": "count",
    "나이": ["mean","min","max"],
    "연봉": ["median", "mean"]
})

print(df3.index)
print(df3.columns)
df3
```

## 문제 1

## 문제 3

```python
df.groupby('genre')['total_sales'].sum().reset_index()
```

## 문제 4

```python
df4 = df.groupby('console')['total_sales'].mean().reset_index()
df4['total_sales'].fillna(0, inplace=True)
df4
```

## 문제 5

```python
df.groupby('developer')['title'].count().reset_index()
```

## 문제 6

```python
cols = [c for c in df.columns if c[-6:] == '_sales']
dict = {}
for col in cols:
    dict.update({col:['mean','sum']})   
df.groupby('console').agg(dict)
```

```python
df.groupby('console').agg({
    'total_sales':['mean','sum'],
    'na_sales':['mean','sum'],
    'jp_sales':['mean','sum'],
    'pal_sales':['mean','sum'],
    'other_sales':['mean','sum']
})
```

## 문제 7

```python
df.groupby('developer').filter(lambda x : x['critic_score'].mean()>9.5)
```

```python
df.groupby('developer')['critic_score'].mean().sort_values(ascending=False)
```
