```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
datasets = [
    "anagrams",
    "anscombe",
    "attention",
    "car_crashes",
    "diamonds",
    "dots",
    "dowjones",
    "exercise",
    "fmri",
    "geyser",
    "glue",
    "healthexp",
    "iris",
    "mpg",
    "penguins",
    "planets",
    "seaice",
    "taxis",
    "tips",
    "titanic"
]

for dataset in datasets:
    df = sns.load_dataset(dataset)
    print(dataset, df.shape)
    display(df.head(3))
    print("\n-----------------------------------")
```

# 수치형 자료

## 연속형 데이터

### 히스토그램

```python
df = sns.load_dataset('mpg')
print(df.shape)
df.head()
```

```python
_=sns.histplot(
    data =df,
    x = 'weight',
    bins = 13,
    color = 'purple'
)
```

### 선 그래프

```python
df = sns.load_dataset('healthexp')
print(df.shape)
df.head()
```

```python
_= sns.lineplot(
    data = df,
    x = 'Year', y='Spending_USD',
    hue = 'Country',
    estimator='sum'
)
```

## 이산형 데이터

### 막대그래프

```python
df = sns.load_dataset('glue')
print(df.shape)
df.head()
```

```python
_=sns.barplot(
    data = df,
    x = 'Task',
    y = 'Score',
#     estimator='mean',
    ci = None
)
```

### 원형그래프

```python
_=df['Year'].value_counts().plot.pie(autopct='%.2f %%')
```

## 상관관계

### 산점도

```python
df = sns.load_dataset('healthexp')
print(df.shape)
df.head()
```

```python
_= sns.scatterplot(
    data = df,
    x = 'Spending_USD', y = 'Life_Expectancy',
    color = 'red'
)
```

### 히트맵

```python
df = sns.load_dataset('car_crashes')
df.drop(columns='abbrev', inplace=True)
print(df.shape)
df.head()
```

```python
corr = df.corr()
display(corr)
_=sns.heatmap(corr, annot=True, cmap='Greys', center=0)
```
