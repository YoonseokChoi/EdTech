```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

# 전처리

```python
df = pd.read_csv("./data/서울시 학원 교습소정보.csv", encoding='euc-kr')
print(df.shape)
df.head()
```

## '개설년도' 컬럼 추가

```python
def f1(x):
    if x!=x:
        return np.nan
    else:
        return int(str(x)[:4])
    
df['개설년도'] = df['등록일자'].apply(f1)

# 2023년도 이전에 개설된 학원만
df = df.loc[df['개설년도']<=2023]
print(df.shape)
df.head()
```

# 서울시내 전체 학원

## 분야별 학원 수

```python
# fig, ax = plt.subplots(figsize=(10,10))
_=df['분야명'].value_counts(normalize=True).plot.pie()
```

## 교습계열별 학원 수

```python
_=df['교습계열명'].value_counts(normalize=True).plot.bar()
```

## 연도별 개설 학원 수

```python
fig, ax = plt.subplots(figsize=(15, 5))

dfdf = df['개설년도'].value_counts().sort_index().to_frame().reset_index()
_=sns.lineplot(data = dfdf,
               x = '개설년도', y='count',
               ax=ax)
_=plt.xticks(rotation=45)
```

# 2020년 이후 개설된 학원

```python
df_20s = df.loc[(df['개설년도']>=2020)&(df['개설년도']<=2023)]
print(df_20s.shape)
df_20s.head()
```

## 교습계열별 학원 수

```python
_=df_20s['교습계열명'].value_counts(normalize=True).plot.bar()
```

# 컴퓨터 학원

```python
com_df = df.loc[df['교습계열명']=="컴퓨터"]

print(com_df.shape)
com_df.head()
```

## 연도별 개설 학원 수

```python
_=com_df['개설년도'].value_counts().sort_index().plot.line()
```

## '구'별 학원 수

```python
_=com_df['행정구역명'].value_counts().plot.bar()
```

# 천재그룹 산하 학원

```python
hb_df = df.loc[(df['학원명'].str.contains('해법'))|(df['학원명'].str.contains('천재'))]
hb_df.reset_index(inplace=True, drop=True)

print(hb_df.shape)
hb_df.head()
```

## '구'별 학원 수

```python
_= hb_df['행정구역명'].value_counts().plot.bar()
```

## 분야별 학원 수

```python
_=hb_df['분야명'].value_counts().plot.bar()
```

## 연도별 개설 학원 수

```python
_=hb_df['개설년도'].value_counts().sort_index().plot.line()
```
