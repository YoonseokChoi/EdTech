## gdp_df (Cities_Brazil_IBGE.xlsx)

```python
gdp_df = pd.read_excel('./data/Cities_Brazil_IBGE.xlsx')
gdp_df['Pib_2014'] = gdp_df['Pib_2014'].str.replace(",",".")
gdp_df['Pib_2014'] = gdp_df['Pib_2014'].astype('float')
gdp_df.rename(columns = {
    'LocalUF':'customer_state'
}, inplace=True)

print(gdp_df.shape)
gdp_df.head(3)
```

```python
gdp_df.groupby(['customer_state'])['Pib_2014'].mean().sort_values().reset_index()
```

### 가설1) 지역별 GDP와 할부개월수는 연관이 있는가?

```python
df1 = df.groupby('customer_state')['payment_installments'].mean().reset_index()
df1.head()
```

```python
state_gdp = gdp_df.groupby('customer_state')['Pib_2014'].mean().sort_values().reset_index()
state_gdp.head()
```

```python
dfdf = pd.merge(df1, state_gdp)
dfdf.head()
```

```python
corr_df = dfdf[['payment_installments', 'Pib_2014']].corr()
corr_df
```

```python
_=sns.heatmap(
    data = corr_df,
    cmap = 'Greys',
    annot = True
)
```

```python
_= sns.lineplot(
    data = dfdf,
    x = 'payment_installments', y = 'Pib_2014',
    color = 'red', ci=None
)
```

> 지역별 GDP와 할부개월 수는 음의 상관관계를 가진다. 

## 지역별 지불액과 GDP의 상관관계

```python
df1 = df.groupby('customer_state')['payment_value'].sum().reset_index()
df1.head()
```

```python
dfdf = pd.merge(df1, state_gdp)
dfdf.head()
```

### 가설2 ) GDP와 지불금액은 상관관계가 있다.

```python
corr_df = dfdf[['payment_value', 'Pib_2014']].corr()
corr_df
```

```python
_= sns.lineplot(
    data = dfdf,
    x = 'payment_value', y = 'Pib_2014',
    color = 'red'
)
```
