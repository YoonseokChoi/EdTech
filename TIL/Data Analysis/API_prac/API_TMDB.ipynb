# import

```python
import requests
from pprint import pprint
import pandas as pd
```

# API key 활용

## Request

```python
api_key = ""

# URL
base_url = "https://api.themoviedb.org/3"
endpoint = "/movie/top_rated"
url = base_url + endpoint

# Parameter
params = {
    'api_key': api_key,
    'language': 'ko-KR'
}

# Header
headers = {'accept':'application/json'}

# Request
response = requests.get(
    url, 
    params = params, 
    headers = headers
)

print(response)
```

## Data 확인

```python
pprint(response.json())
```

```python
df = pd.DataFrame(response.json()['results'])

print(df.shape)
df.head()
```

# 예제 1 - 영화 제목 검색기

```python
def search_mv(query):
    api_token = ""
    
    # URL
    base_url = "https://api.themoviedb.org/3"
    endpoint = "/search/movie"
    url = base_url + endpoint

    # Parameter
    params = {
        'query': query,
        'language': 'ko-KR'
    }

    # Header
    headers = {
        "Authorization": f"Bearer {api_token}",
        "accept": "application/json"
    }

    # Request
    response = requests.get(
        url, 
        params = params, 
        headers = headers
    )
    
    return pd.DataFrame(response.json()['results']) 
    
query = input("영화 이름을 입력하삼 : ")    
df = search_mv(query)
df.head()
```

## 1~5페이지까지 요청

```python
api_token = ""

# URL
base_url = "https://api.themoviedb.org/3"
endpoint = f"/movie/now_playing"
url = base_url + endpoint

response_data = []
for i in range(1,6):
    # Parameter
    params = {'page': f"{i}"}

    # Header
    headers = {
        "Authorization": f"Bearer {api_token}",
        "accept": "application/json"
    }

    # Request
    response = requests.get(
        url, 
        params = params,
        headers = headers
    )
    
    print(response)
    response_data.extend(response.json()['results'])
```

```python
df = pd.DataFrame(response_data)
print(df.shape)
df.head()
```

## 평균 평점(vote_average)이 가장 높은 영화의 리뷰 데이터 요청, 출력

```python
top_mv_df = df.loc[df['vote_average'] == df['vote_average'].max()]['id']
top_mv_id = top_mv_df.values[1]
print(top_mv_id)
```

```python
api_token = ""
movie_id = top_mv_id
# movie_id = top_id

# URL
base_url = "https://api.themoviedb.org/3"
endpoint = f"/movie/{movie_id}/reviews"
url = base_url + endpoint

# Parameter
# params = {'page': ''}

# Header
headers = {
    "Authorization": f"Bearer {api_token}",
    "accept": "application/json"
}

# Request
response = requests.get(
    url, 
    headers = headers
)

print(response)
```

```python
for i, r in enumerate(response.json()['results']):
    print(f"[{i+1}]\n\n{r['content']}\n")
    print("-----------")
```
