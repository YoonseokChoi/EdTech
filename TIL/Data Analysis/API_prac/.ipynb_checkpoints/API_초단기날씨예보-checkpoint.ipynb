```python
import requests
from pprint import pprint

url = 'http://apis.data.go.kr/1360000/VilageFcstInfoService_2.0/getUltraSrtFcst'
params ={
    'serviceKey' : '', 
    'pageNo' : '1', 
    'numOfRows' : '1000', 
    'dataType' : 'JSON', 
    'base_date' : '20241216', 
    'base_time' : '1100', 
    'nx' : '61', 
    'ny' : '125' }

response = requests.get(url, params=params)
# print(response.content)
response_python = response.json()
pprint(response_python)
```

```python
item = response_python['response']['body']['items']['item']
pprint(item)
```

## 습도 출력

```python
for i in item:
    if i['category'] == 'REH':
        print(i['category'], i['fcstValue'])
        # print(i)
```

## 온도 출력

```python
for i in item:
    if i['category'] == 'T1H':
        print(i['category'], i['fcstValue'])
        # print(i)
```

## 최고, 최저 온도 출력

```python
tmps = []
for i in item:
    if i['category'] == 'T1H':
        # print(i['category'], i['fcstValue'])
        tmps.append(int(i['fcstValue']))
   
print(tmps)
```

```python
for i in item:
    if i['category'] == 'T1H' and i['fcstValue'] == str(max(tmps)):
        print(i)
```

```python
for i in item:
    if i['category'] == 'T1H' and i['fcstValue'] == str(min(tmps)):
        print(i)
```
