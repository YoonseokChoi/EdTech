```python
import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd
import numpy as np
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

_ = plt.plot(x,y, label='y')
_= plt.xlabel('X축')
_= plt.ylabel('Y축', rotation=0)
_= plt.legend()
plt.show()
```

# scatterplot

```python
df = sns.load_dataset("tips")
print(df.shape)
df.head()
```

```python
_=sns.scatterplot(
    data = df, 
    x ='total_bill', y='tip',
    color = 'purple'
)

# _=plt.xlabel('총 지불액', fontsize=8)
# _=plt.ylabel('TIP', rotation=0, fontsize=8)
# _=plt.title("< 제목 >", fontsize=12)

plt.show()
```

```python
_=sns.scatterplot(
    data = df, 
    x ='total_bill', y='tip', hue = 'smoker'
)

_=plt.xlabel('x축이다', fontsize=8)
_=plt.ylabel('y축이다', rotation=0, fontsize=8)
_=plt.title("< 제목이다 >", fontsize=12)

plt.show()
```
